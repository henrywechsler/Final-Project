```python
import requests
import csv 
import json
import datetime
import pandas as pandasForSortingCSV
api_key = 'v4HTRgnYY94GBteg3f2py73d3wGVDgMD4sgLZtWP'
```

```python
def display_NBA_standings():
    url = f"https://api.sportradar.com/nba/trial/v8/en/seasons/2023/REG/standings.json?api_key={api_key}"
    csv_file = open("NBAStandings.csv", "w", newline = "", encoding = "utf-8")
    csv_writer = csv.writer(csv_file)
    csv_writer.writerow(["Name", "Conference", "Wins", "Losses", "Win Pct", "GB", "PPG", "PAPG", "PD"])
    standings = requests.get(url).json()
    confNum = 0
    divNum = 0
    teamNum = 0
    while confNum < 2:
        conference = standings["conferences"][confNum]
        while divNum < 3:
            division = conference["divisions"][divNum]
            while teamNum < 5:
                team = division["teams"][teamNum]
                team_name = team["name"]
                if confNum == 0:
                    team_conf = "West"
                else:
                    team_conf = "East"
                team_w = team["wins"]
                team_l = team["losses"]
                team_pct = team["win_pct"]
                team_gb = team["games_behind"]["conference"]
                team_ppg = team["points_for"]
                team_papg = team["points_against"]
                team_pd = team["point_diff"]
                csv_writer.writerow([team_name, team_conf, team_w, team_l, team_pct, team_gb, team_ppg, team_papg, team_pd])
                teamNum += 1
            divNum += 1
            teamNum = 0
        confNum += 1
        divNum = 0
    csv_file.close()
    sheet = pandasForSortingCSV.read_csv("NBAStandings.csv")
    sheet.sort_values(["Win Pct"], ascending=[False],inplace=True)
    return sheet
display_NBA_standings()
```

```python
import requests
import tkinter as tk
from tkinter import ttk


def display_NBA_standings():
    api_key = 'v4HTRgnYY94GBteg3f2py73d3wGVDgMD4sgLZtWP'
    url = f"https://api.sportradar.com/nba/trial/v8/en/seasons/2023/REG/standings.json?api_key={api_key}"
    standings = requests.get(url).json()
    confNum = 0
    divNum = 0
    teamNum = 0
    root = tk.Tk()
    treeview = ttk.Treeview(columns=("Conference", "Wins", "Losses", "Win Pct", "GB", "PPG", "PAPG", "PD"))
    while confNum < 2:
        conference = standings["conferences"][confNum]
        while divNum < 3:
            division = conference["divisions"][divNum]
            while teamNum < 5:
                team = division["teams"][teamNum]
                team_name = team["name"]
                if confNum == 0:
                    team_conf = "West"
                else:
                    team_conf = "East"
                team_w = team["wins"]
                team_l = team["losses"]
                team_pct = team["win_pct"]
                team_gb = team["games_behind"]["conference"]
                team_ppg = team["points_for"]
                team_papg = team["points_against"]
                team_pd = team["point_diff"]        
                treeview.heading("#0", text="Team")
                treeview.heading("Conference", text="Confrence")
                treeview.heading("Wins", text="Wins")
                treeview.heading("Losses", text="Losses")
                treeview.heading("Win Pct", text="Win Pct")
                treeview.heading("GB", text="GB")
                treeview.heading("PPG", text="PPG")
                treeview.heading("PAPG", text="PAPG")
                treeview.heading("PD", text="PD")
                treeview.insert('',tk.END, text=team_name, values=(team_conf, team_w, team_l, team_pct, team_gb, team_ppg, team_papg, team_pd))
                teamNum += 1
            divNum += 1
            teamNum = 0
        confNum += 1
        divNum = 0
    

    treeview.pack()
    root.mainloop()
display_NBA_standings()
```

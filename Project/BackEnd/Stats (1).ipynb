```python
def points_leaders():
    url = f"https://api.sportradar.com/nba/trial/v8/en/seasons/2023/REG/leaders.json?api_key={api_key}"
    stats = requests.get(url).json()
    root = tk.Tk()
    columns=("Name", "Pos", "Team", "PPG", "TS%")
    treeview = ttk.Treeview(root, height="500",columns=columns)
    for i in stats["categories"]:
         if i["name"] == "points" and i["type"] == "average":
            leaders = i["ranks"]
            for i in leaders:
                rank = i["rank"]
                name = i["player"]["full_name"]
                pos = i["player"]["primary_position"]
                team = i["teams"][0]["name"]
                ppg = i["score"]
                ts = i["total"]["true_shooting_pct"]
                treeview.heading("#0", text="Rank")
                treeview.heading("Name", text="Name")
                treeview.heading("Pos", text="Pos")
                treeview.heading("Team", text="Team")
                treeview.heading("PPG", text="PPG")
                treeview.heading("TS%", text="TS%")
                treeview.insert('',tk.END, text=rank, values=(name, pos, team, ppg, ts))
    treeview.pack()
    root.mainloop()
points_leaders()
```

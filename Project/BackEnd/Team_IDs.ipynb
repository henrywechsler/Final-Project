```python
def team_id():
    teams_dict = {}
    url = f"https://api.sportradar.com/nba/trial/v8/en/seasons/2023/REG/standings.json?api_key={api_key}"
    standings = requests.get(url).json()
    csv_file = open("teams.csv", "w", newline = "", encoding = "utf-8")
    csv_writer = csv.writer(csv_file)
    csv_writer.writerow(["ID", "Team"])
    confNum = 0
    divNum = 0
    teamNum = 0
    while confNum < 2:
        conference = standings["conferences"][confNum]
        while divNum < 3:
            division = conference["divisions"][divNum]
            while teamNum < 5:
                team = division["teams"][teamNum]
                team_name = team["name"]
                team_id = team["id"]
                teams_dict[team_name] = team_id
                teamNum += 1
            divNum += 1
            teamNum = 0
        confNum += 1
        divNum = 0
    return teams_dict
team_id()
```

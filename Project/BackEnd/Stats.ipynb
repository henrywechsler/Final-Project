```python
import requests
import csv 
import json
import datetime
import pandas as pandasForSortingCSV
api_key = 'v4HTRgnYY94GBteg3f2py73d3wGVDgMD4sgLZtWP'
```

```python
def points_leaders():
    url = f"https://api.sportradar.com/nba/trial/v8/en/seasons/2023/REG/leaders.json?api_key={api_key}"
    csv_file = open("NBAStatsPoints.csv", "w", newline = "", encoding = "utf-8")
    csv_writer = csv.writer(csv_file)
    csv_writer.writerow(["Rank", "Name", "Pos", "Team", "PPG", "TS%"])
    stats = requests.get(url).json()
    for i in stats["categories"]:
         if i["name"] == "points" and i["type"] == "average":
            leaders = i["ranks"]
            for i in leaders:
                rank = i["rank"]
                name = i["player"]["full_name"]
                pos = i["player"]["primary_position"]
                team = i["teams"][0]["name"]
                ppg = i["score"]
                ts = i["total"]["true_shooting_pct"]
                csv_writer.writerow([rank, name, pos, team, ppg, ts])
    csv_file.close()
    sheet = pandasForSortingCSV.read_csv("NBAStatsPoints.csv")
    return sheet
points_leaders()
```
